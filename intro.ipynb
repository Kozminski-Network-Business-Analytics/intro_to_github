```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create figure and axis
fig, ax = plt.subplots(figsize=(6, 6))

# Add a square
square = patches.Rectangle((0.1, 0.1), 0.3, 0.3, fill=False, linewidth=3)
ax.add_patch(square)

# Add a circle
circle = patches.Circle((0.7, 0.7), 0.15, fill=False, linewidth=3)
ax.add_patch(circle)

# Add a triangle
triangle = patches.Polygon([[0.6, 0.1], [0.9, 0.1], [0.75, 0.35]],
                           fill=False, linewidth=3)
ax.add_patch(triangle)

# Set limits and show plot
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
plt.title("Simple Shapes in Python")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(6, 6))

# --- House Body (Square) ---
house_body = patches.Rectangle((0.3, 0.2), 0.4, 0.4, fill=False, linewidth=3)
ax.add_patch(house_body)

# --- Roof (Triangle) ---
roof = patches.Polygon([[0.25, 0.6], [0.75, 0.6], [0.50, 0.85]],
                       fill=False, linewidth=3)
ax.add_patch(roof)

# --- Door ---
door = patches.Rectangle((0.45, 0.2), 0.1, 0.2, fill=False, linewidth=2)
ax.add_patch(door)

# --- Windows ---
window1 = patches.Rectangle((0.35, 0.45), 0.1, 0.1, fill=False, linewidth=2)
window2 = patches.Rectangle((0.55, 0.45), 0.1, 0.1, fill=False, linewidth=2)
ax.add_patch(window1)
ax.add_patch(window2)

# --- Finalize ---
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
plt.title("Simple House Shape in Python")
plt.show()
```
